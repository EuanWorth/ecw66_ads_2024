```python
fetch_radius = 15
small_search_radius = 1
large_search_radius = 5
```

```python
queue = {oa for oa in oa_list["oa21cd"]}
already_done = {oa21cd[0] for oa21cd in fynesse.access.sql_select(conn, "SELECT oa21cd FROM polygon_counts;")}
print(f"Processed {len(already_done)}/{len(oa_list)} ({int(100*len(already_done)/len(oa_list))}%) already")
for item in already_done:
  if item in queue:
    queue.remove(item)

while len(queue) > 0:
  oa = queue.pop()
  print(f"Processing {oa}")
  latitude, longitude = (fynesse.access.sql_select(conn, f"SELECT latitude, longitude FROM oa_geo_data WHERE oa21cd = '{oa}';"))[0]
  degrees_per_vkm, degrees_per_hkm = fynesse.access.calc_degrees_per_km(latitude)
  lower_longitude_bound = longitude - (fetch_radius - large_search_radius) * degrees_per_hkm
  upper_longitude_bound = longitude + (fetch_radius - large_search_radius) * degrees_per_hkm
  lower_latitude_bound = latitude - (fetch_radius - large_search_radius) * degrees_per_vkm
  upper_latitude_bound = latitude + (fetch_radius - large_search_radius) * degrees_per_vkm
  oa_geo_data = fynesse.access.sql_select(conn, f"SELECT * FROM oa_geo_data WHERE latitude >= {lower_latitude_bound} AND latitude <= {upper_latitude_bound} AND longitude >= {lower_longitude_bound} AND longitude <= {upper_longitude_bound};")
  oa_df = (pd.DataFrame(oa_geo_data, columns=["fid", "oa21cd", "lsoa21cd", "lsoa21nm", "lsoa21nmw", "BNG_E", "BNG_N", "latitude", "longitude", "globalid", "geometry"]))
  oa_df['geometry'] = gpd.GeoSeries.from_wkt(oa_df['geometry'])
  oa_gdf = gpd.GeoDataFrame(oa_df, geometry='geometry')
  print(f"Fetched {len(oa_gdf)} oas")
  lower_fetch_longitude_bound = longitude - (fetch_radius) * degrees_per_hkm
  upper_fetch_longitude_bound = longitude + (fetch_radius) * degrees_per_hkm
  lower_fetch_latitude_bound = latitude - (fetch_radius) * degrees_per_vkm
  upper_fetch_latitude_bound = latitude + (fetch_radius) * degrees_per_vkm
  osm_gdf = ox.features_from_bbox(tags=tags, bbox=(lower_longitude_bound, lower_latitude_bound, upper_longitude_bound, upper_latitude_bound))
  print(f"Fetched {len(osm_gdf)} osm features")
  polygon_counts = {}
  small_radius_counts = {}
  large_radius_counts = {}
  percent = 0
  count = 0
  for _, row in oa_gdf.iterrows():
    if int((count / len(oa_gdf)) * 10) > percent:
      percent = int((count / len(oa_gdf)) * 10)
      print(f"{percent}0% complete")
    count += 1
    if row["oa21cd"] not in queue:
      if row["oa21cd"] != oa:
        continue
    else:
      queue.remove(row["oa21cd"])
    large_xmin = row["longitude"] - (large_search_radius) * degrees_per_hkm
    large_xmax = row["longitude"] + (large_search_radius) * degrees_per_hkm
    large_ymin = row["latitude"] - (large_search_radius) * degrees_per_vkm
    large_ymax = row["latitude"] + (large_search_radius) * degrees_per_vkm
    small_xmin = row["longitude"] - (small_search_radius) * degrees_per_hkm
    small_xmax = row["longitude"] + (small_search_radius) * degrees_per_hkm
    small_ymin = row["latitude"] - (small_search_radius) * degrees_per_vkm
    small_ymax = row["latitude"] + (small_search_radius) * degrees_per_vkm
    large_polygon = shapely.geometry.box(large_xmin, large_ymin, large_xmax, large_ymax)
    small_polygon = shapely.geometry.box(small_xmin, small_ymin, small_xmax, small_ymax)
    exact_polygon = row["geometry"]
    large_gdf = (osm_gdf[osm_gdf['geometry'].within(large_polygon)])
    small_gdf = (large_gdf[large_gdf['geometry'].within(small_polygon)])
    exact_gdf = (osm_gdf[osm_gdf['geometry'].within(exact_polygon)])
    polygon_counts[row["oa21cd"]] = fynesse.access.count_pois(exact_gdf, feature_list)
    small_radius_counts[row["oa21cd"]] = fynesse.access.count_pois(small_gdf, feature_list)
    large_radius_counts[row["oa21cd"]] = fynesse.access.count_pois(large_gdf, feature_list)
  polygon_df = pd.DataFrame.transpose(pd.DataFrame.from_dict(polygon_counts))
  polygon_df.to_csv(f"polygon_counts.csv")
  %sql LOAD DATA LOCAL INFILE "polygon_counts.csv" INTO TABLE `polygon_counts` FIELDS TERMINATED BY ',' OPTIONALLY ENCLOSED BY '"' LINES TERMINATED BY '\n' IGNORE 1 LINES;
  small_radius_df = pd.DataFrame.transpose(pd.DataFrame.from_dict(small_radius_counts))
  small_radius_df.to_csv(f"small_radius_counts.csv")
  %sql LOAD DATA LOCAL INFILE "small_radius_counts.csv" INTO TABLE `small_radius_counts` FIELDS TERMINATED BY ',' OPTIONALLY ENCLOSED BY '"' LINES TERMINATED BY '\n' IGNORE 1 LINES;
  large_radius_df = pd.DataFrame.transpose(pd.DataFrame.from_dict(large_radius_counts))
  large_radius_df.to_csv(f"large_radius_counts.csv")
  %sql LOAD DATA LOCAL INFILE "large_radius_counts.csv" INTO TABLE `large_radius_counts` FIELDS TERMINATED BY ',' OPTIONALLY ENCLOSED BY '"' LINES TERMINATED BY '\n' IGNORE 1 LINES;
  done = %sql SELECT COUNT(*) FROM polygon_counts
  print(f"Done {done[0][0]} / {len(oa_list)}")

```

In the access stage I initially downloaded the relavent census data (TS062 and TS011) and uploaded them to the AWS DB.

Then I found the geometry data and uploaded that to the notebook, and extracted the data to a gdf using geopandas. This data was then uploaded to the database. This took multiple tries, since initially the database was configured with sufficient space for the geometry data, which could be over 64KiB per cell. This meant that the data was truncated and the geometry was lost. This had another issue, when attempting to match a test data point with OSM data the data crashed. This was because the census is built on the BGN CRS. This meant that the data needed to be remapped and reuploaded so that it was in the same CRS as the OSM data.

The final part was to fetch and associate the OSM data with the OA geometries. For this I had decided to fetch, aggregate and then upload to the data, in an attempt to create something more useful. I initially selected a set of OSM features that I thought would be a set of initial features to use for prediction.  selected that for each OA we would save the features within the OA geometry itself, within 1km of it's center point and within 5km of the center point. This would allow us to see the impact of both the OA and it's local area. The challenge here was trading off between the size of data and the time to make an API call. The google collab instance would run out of RAM if you tried to fetch all the features for too large an area (~40km square in London). The solution here was to do a mixture. The code would select an OA, take a large square around it. It would find all of the other OAs within this square (not too close to the edge) and fetch their geometries. It then fetched all of the features in the large square and checked which OAs then could be associated with. It then uploaded all this aggregated data to the database

```python
deprivation_design_matrix = deprivation_df[["large_fuel", "exact_books", "small_bank", "exact_cafe", "exact_university"]]
deprivation_design_matrix["large_sport_normalised"] = deprivation_df["large_sport"] / deprivation_df["large_sum"]
deprivation_design_matrix["small_historic_normalised"] = deprivation_df["small_historic"] / deprivation_df["small_sum"]
deprivation_design_matrix["small_restaurant_normalised"] = deprivation_df["small_restaurant"] / deprivation_df["small_sum"]
deprivation_design_matrix["small_convenience_normalised"] = deprivation_df["small_convenience"] / deprivation_df["small_sum"]
deprivation_design_matrix["large_university_normalised"] = deprivation_df["large_university"] / deprivation_df["large_sum"]
deprivation_design_matrix["log_small_school"] = np.log10(deprivation_df["small_school"] + 1)
deprivation_design_matrix["log_small_historic"] = np.log10(deprivation_df["small_historic"] + 1)
deprivation_design_matrix["log_sport_exact"] = np.log10(deprivation_df["exact_sport"] + 1)
deprivation_design_matrix.set_index(deprivation_df.index, inplace=True)
deprivation_design_matrix = sm.add_constant(deprivation_design_matrix)
deprivation_design_matrix = deprivation_design_matrix.astype(float)
deprivation_design_matrix
```

```python
deprivation_model = sm.OLS(deprivation_df["average_deprivation"].astype(float), deprivation_design_matrix)
deprivation_results = deprivation_model.fit()
print(deprivation_results.summary())
```
